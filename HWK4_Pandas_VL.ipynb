```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "C://Users//Lucas//Desktop//schools_complete.csv"
student_data_to_load = "C://Users//Lucas//Desktop//students_complete.csv"
```

```python
# Read School Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
school_data.head ()
```

```python
# Read Student Data File and store into Pandas Data Frames
student_data = pd.read_csv(student_data_to_load)
student_data.head ()
```

```python
# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head ()
```

```python
school_data_complete_df.columns
```

```python
#Find unique number of schools
unique_schools = school_data_complete_df["school_name"].unique()
unique_schools
```

```python
#Calculate the total number of schools
school_count = len(school_data_complete_df["school_name"].unique())
school_count
```

```python
#Calculate the total number of students
student_count = len(school_data_complete_df["student_name"])
student_count
```

```python
#Calculate the total budget
total_budget = school_data_complete_df["budget"].sum()
total_budget
```

```python
#Calculate the average math score 
math_avg = school_data_complete_df["math_score"].mean()
math_avg
```

```python
#Calculate the average reading score
reading_avg = school_data_complete_df["reading_score"].mean()
reading_avg
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_avg=(math_avg+reading_avg)/2
overall_avg
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
number_passing_math = school_data_complete_df.loc[school_data_complete_df["math_score"] > 69, [
        "student_name", "math_score", "school_name"]]
number_passing_math.head()
```

```python
passing_math_count = len(number_passing_math["student_name"])
passing_math_count
```

```python
pass_math=(passing_math_count/student_count)*100
pass_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
number_passing_reading = school_data_complete_df.loc[school_data_complete_df["reading_score"] > 69, [
        "student_name", "reading_score", "school_name"]]
number_passing_reading.head()
```

```python
passing_reading_count = len(number_passing_reading["student_name"])
passing_reading_count
```

```python
pass_reading=(passing_reading_count/student_count)*100
pass_reading
```

```python
#Create a dataframe to hold the above results
# Place all of the data found into a summary DataFrame

district_summary = pd.DataFrame({"Total Students": [student_count],
                              "Total Budget": [total_budget],
                              "Average Math Score": [math_avg],
                              "Average Reading Score": [reading_avg],
                              "%Passing Math": [pass_math],
                              "%Passing Reading": [pass_reading],
                              "%Overall Passing Rate": [overall_avg]})
district_summary
```
